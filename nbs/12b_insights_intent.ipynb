# Insights Intent
> Classify query intent based on keyword patterns.

```python
#| export
from sqlmodel import Session
from seootter.gsc.queries import get_top_queries
```

```python
# | export
INTENT_PATTERNS = {
    "informational": [
        "what is",
        "how to",
        "guide",
        "explain",
        "tutorial",
        "ما هو",
        "كيف",
        "شرح",
        "تعريف",
        "دليل",
    ],
    "comparison": [
        "vs",
        "alternative",
        "best",
        "compare",
        "difference",
        "مقارنة",
        "أفضل",
        "بديل",
        "الفرق بين",
        "ضد",
    ],
    "transactional": [
        "buy",
        "price",
        "review",
        "specs",
        "discount",
        "سعر",
        "شراء",
        "مراجعة",
        "مواصفات",
        "طلب",
    ],
    "commercial": [
        "pricing",
        "cheapest",
        "affordable",
        "top rated",
        "comparison",
        "worth it",
        "cost",
        "budget",
        "rental",
        "أسعار",
        "أرخص",
        "تكلفة",
        "ميزانية",
        "إيجار",
    ],
    "navigational": [
        "site:",
        "login",
        "official",
        "website",
        "github",
        "موقع",
        "تسجيل دخول",
        "الرسمي",
    ],
}
```

```python
# | export
def classify_intent(query: str
                    ) -> str:
    "Classify query intent based on keyword patterns."
    for intent, patterns in INTENT_PATTERNS.items():
        if any(p in query.lower() for p in patterns): return intent
    return "informational"
```

```python
# | export
def classify_page_intents(session: Session,
                          site_url: str,
                          start_date: str,
                          end_date: str,
                          country: str | None = None,
                          page_path: str | None = None,
                          limit: int = 10
                          ) -> list[dict]:
    "Get top queries with intent classification."
    return [{**row, "intent": classify_intent(row["query"])}
            for row in get_top_queries(session, site_url, start_date, end_date,
                                       country=country, page_path=page_path, limit=limit)]
```
